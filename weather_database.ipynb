```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
```

```python
#generate 2000 random longitudes and latitudes
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
#create a list of the random lngs/lats
coordinates = list(lat_lngs)
```

```python
#create an empty list of cities
cities = []
#run a for loop to find the nearest city to each random coordinate and then add it to the list above
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    #make sure we do not add the same city more than once with if statement
    if city not in cities:
        cities.append(city)
        
len(cities)
```

```python
#begin creating api call url with our weather api key
from config import weather_api_key
import time
from datetime import datetime
import requests 
```

```python
#create a url with our api key from the openweathermap documentation
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&appid=" + weather_api_key
```

```python
#run a test url to inspect the api and understand what the data looks like
boston_url = url + "&q=Boston"
print(boston_url)
```

```python
# first create an empty list to hold our city data
weather_data = []
#print the beginning of data detrievel
print("Beginning Data Retrievel")
print("----------------------")

#create counters
record_count = 1
set_count = 1
```

```python
#create a for loop to loop through cities
for i, city in enumerate(cities):
    
    if (i % 50 == 0 and i >=50):
        set_count += 1
        record_count = 1
        time.sleep(60)
        
    city_url = url + "&q=" + cities[i]
    
    print(f"Processing record {record_count} of Set {set_count} | {city}")
    
    record_count += 1
    # Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_desc = city_weather["weather"][0]["description"]
        city_country = city_weather["sys"]["country"]
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into weather_data list.
        weather_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Weather Description": city_desc,
                          "Country": city_country,
                          "Date": city_date})

# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
weather_data_df = pd.DataFrame(weather_data)
weather_data_df.head()
```

```python
#create a csv file to hold the new dataframe
output_data_file = "weather_database/weatherpy_database.csv"

#export data to file we created above
weather_data_df.to_csv(output_data_file, index_label="City_ID")
```
